```python
import numpy as np
import pandas as pd

df = pd.read_csv("students.csv")
print("Student Data:")
print(df)

math_mean = np.mean(df["Math"])
science_mean = np.mean(df["Science"])
english_mean = np.mean(df["English"])

print("\nSubject Averages:")
print("Math Mean:", math_mean)
print("Science Mean:", science_mean)
print("English Mean:", english_mean)

marks_array = np.array(df.iloc[:3, 1:])
reshaped = marks_array.reshape(3, 3)

print("\nReshaped Marks Array (3 students x 3 subjects):")
print(reshaped)

average_series = pd.Series({
    "Math": math_mean,
    "Science": science_mean,
    "English": english_mean
})

print("\nAverage Marks Series:")
print(average_series)

df["Total"] = df["Math"] + df["Science"] + df["English"]
print("\nDataFrame with Total Marks:")
print(df)
```
