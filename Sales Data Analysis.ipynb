```python
import pandas as pd

# Load dataset
df = pd.read_csv("sales_data.csv")

print("=== Sales Dataset ===")
print(df)

# 1. Filter with conditions
print("\n=== Sales greater than 1000 ===")
high_sales = df[df['Sales'] > 1000]
print(high_sales)

# 2. Sorting
print("\n=== Sorted by Sales (Descending) ===")
sorted_sales = df.sort_values(by='Sales', ascending=False)
print(sorted_sales)

# 3. Aggregation
print("\n=== Total Sales by Region ===")
sales_by_region = df.groupby('Region')['Sales'].sum()
print(sales_by_region)

print("\n=== Average Sales by Product ===")
avg_sales_product = df.groupby('Product')['Sales'].mean()
print(avg_sales_product)

# 4. Descriptive Statistics
print("\n=== Sales Statistics ===")
print(df['Sales'].describe())

print("\n=== Product Count ===")
print(df['Product'].value_counts())
```
