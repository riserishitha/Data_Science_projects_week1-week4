```python
import math
import random

print("‚≠ê Welcome to the Math Quiz Game ‚≠ê")
score = 0

for i in range(5):
    num1 = random.randint(1, 20)
    num2 = random.randint(1, 20)
    operation = random.choice(['add', 'subtract', 'multiply', 'square_root'])

    if operation == 'add':
        correct_answer = num1 + num2
        user_answer = int(input(f"{i+1}. What is {num1} + {num2}? "))

    elif operation == 'subtract':
        correct_answer = num1 - num2
        user_answer = int(input(f"{i+1}. What is {num1} - {num2}? "))

    elif operation == 'multiply':
        correct_answer = num1 * num2
        user_answer = int(input(f"{i+1}. What is {num1} * {num2}? "))

    else:
        num = random.randint(1, 30)
        correct_answer = round(math.sqrt(num), 2)
        user_answer = float(input(f"{i+1}. What is the square root of {num} (round to 2 decimals)? "))

    if user_answer == correct_answer:
        print("‚úîÔ∏è Correct!\n")
        score += 1
    else:
        print(f"‚ùå Wrong! Correct answer: {correct_answer}\n")

print(f"üéâ Your final score: {score}/5")
```
