```python
def add(a, b):
    return a + b

def largest(a, b):
    if a > b:
        return a
    else:
        return b

def is_even(n):
    if n % 2 == 0:
        return True
    else:
        return False

def square(n):
    return n * n

print("Function Utility Tool")
print("1. Add two numbers")
print("2. Find largest number")
print("3. Check even or odd")
print("4. Find square")

choice = int(input("Enter your choice: "))

if choice == 1:
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))
    print("Sum =", add(a, b))

elif choice == 2:
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))
    print("Largest =", largest(a, b))

elif choice == 3:
    n = int(input("Enter number: "))
    if is_even(n):
        print(n, "is Even")
    else:
        print(n, "is Odd")

elif choice == 4:
    n = float(input("Enter number: "))
    print("Square =", square(n))

else:
    print("Invalid choice!")
```
